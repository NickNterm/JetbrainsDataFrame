```kotlin
%useLatestDescriptors
%use dataframe
import kotlin.random.Random

var df = @Import DataFrame.read("netflix.csv").add("Score") { Random.nextInt(0, 6) }
//df
```

```kotlin
df = df.add("Decade") {
    `Release Year` / 10 * 10
}
df = df.rename("Release Year").into("ReleaseYear")
df
```

```kotlin
// Here We know that score is
var maxScore = df.max { Score }
var minScore = df.min { Score }

fun getColor(min: Int, max: Int, value: Int): String {
    require(min <= max) { "min must be less than or equal to max" }
    require(value in min..max) { "value must be between min and max" }

    val ratio = (value - min).toFloat() / (max - min).toFloat()
    val red = ((1 - ratio) * 255).toInt()
    val green = (ratio * 255).toInt()
    val blue = 0 // Keep blue constant for a red-to-green gradient

    return String.format("#%02X%02X%02X", red, green, blue)
}

df = df.add("ScoreColor") { getColor(minScore, maxScore, Score) }
df
```

```kotlin
df = df.add("isSeries") { Duration.endsWith("min") }
df
```

```kotlin
df.writeCSV("output.csv")
```
